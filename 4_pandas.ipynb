## Demo funciones pandas

```python
%pylab inline
# importando pandas usando la convenci√≥n
import pandas as pd
```

```python
# Cargar base de datos imdb de titulos de peliculas
titulos = pd.read_csv('https://public-access-s3.s3.amazonaws.com/titles.csv' )
titulos.head()
```

```python
# Cargar base de datos de elenco de peliculas, de imdb
elenco = pd.read_csv('https://public-access-s3.s3.amazonaws.com/cast.csv', encoding='utf-8')
elenco.head()
```

```python
# Cuantas peliculas estan listadas en el dataframe de titulos?
len(titulos)
```

### Revisaremos comandos basicos pandas, sobre estas series

```python
# Cuales son las peliculas mas antiguas listadas en titulos?
titulos.sort_values('year').head()
```

### Ejercicios

```python
# Cuantas peliculas estan listadas por el nombre "Hermosa" ?
len(titulos[titulos.title== "Hermosa"])
```

```python
# Titulos mas comunes en la historia filmografica
titulos.title.value_counts().head(50)
```

```python
# Cual fue la primer pelicula hecha titulada "Maya" 
titulos[titulos.title == "Maya"].sort_values('year').head(1)
```

```python
# Listar todas las peliculas que contengan la palabra "Jesus" ordenadas de la mas vieja a la mas reciente.
titulos[titulos.title.str.contains('Jesus')].sort_values('year')
```

```python
# Cuantas peliculas fueron hechas en el a√±o 2000?
len(titulos[titulos.year== 2000])
```

```python
# Cuantas peliculas fueron hechas en el a√±o 1970?
len(titulos[titulos.year==1970])
```

```python
# Cuantas peliculas fueron hechas de 1980 a 1990
len(titulos[(titulos.year>=1980) & (titulos.year<=1990)].sort_values('year'))
```

```python
# Peliculas hechas en la decada de los 80
titulos[titulos.year//10 == 198]
```

```python
# En que a√±os alguna pelicula llamada "Maya" se presento
titulos[ titulos.title == "Maya"].sort_values('year')
```

```python
# Cuantos roles o papeles hubo en la pelicula "Maya"
len(elenco[elenco.title == "Maya"].sort_values('name'))
```

```python
# Cuantos papeles en la pelicula "The Maya" no estan clasificados en algun valor "n"

nan = elenco[elenco.title=="Maya"]
nan= nan[nan.n.isnull()]
len(nan)
```

```python
# Cuantos papeles en la pelicula "A clockwork orange" si estuvieron clasificados con un valor "n"
#elenco[elenco.title.str.contains("clockwork orange")]
#No existe la pelicula en la lista, pero sacamos de los que contengan la palabra "Orange"
orange = elenco[elenco.title.str.contains("orange")]
orange = orange[orange.n.notnull()]
len(orange)
```

```python
# Mostrar el elenco completo de la pelicula "Batman" 
# ordenado por su clasificasi√≥n "n", ignorando los papeles que 
# no se les asigno ningun valor "n"
batman = elenco[elenco.title == "Batman"]
batman = batman[batman.n.notnull()].sort_values('n').head(50)
batman
```

```python
# Mostrar el elenco completo ordenado por la clasificacion "n"
# de la pelicula "Batman" de 1989
bat_1989 = elenco[(elenco.title == 'Batman') & (elenco.year == 1989)]
bat_1989 = bat_1989[bat_1989.n.notnull()].sort_values('n').head(50)
bat_1989
```

```python
# Cuantos papeles fueron listados en la pelicula "Batman" de 1989
bat_99 = elenco[(elenco.title=='Batman')&(elenco.year== 1989)]
bat_99
```

```python
# Cuantos papeles de "John Smith" han sido hechos en la historia 
# de las peliculas

john = elenco[elenco.character == 'John Smith']
len(john)
```

```python
# Cuanta gente ha hecho el papel de "Robin"
robin = elenco[elenco.character== 'Robin']
len(robin)
```

```python
# Cuantos papeles ha hecho "Selena Gomez" en su carrera
selena = elenco[elenco.name== 'Selena Gomez']
selena
```

```python
# Listado de papeles de soporte que tuvo el actor "Charlton Heston" en
# la decada de los 50's, ordenado por a√±o

charl = elenco[(elenco.name=='Charlton Heston') & (elenco.n == 2) & (elenco.year// 10 == 195)]
charl 
```

```python
# Listado de papeles como protagonista que tuvo el actor "Charlton Heston" en
# la decada de los 60's, ordenado por a√±o de forma descendente


charlton = elenco[elenco.name == "Charlton Heston"]
charlton = charlton[charlton.n == 1]
charlton = charlton[charlton.year// 10 == 196]
charlton = charlton.sort_values('year', ascending = False)
charlton
```

```python
# ¬øCuantos papeles para actores hubo en la decada de los 60's?

actores = elenco[(elenco.type == 'actor')&(elenco.year//10 == 196)]
len(actores)
```

```python
# ¬øCuantos papeles para actrices hubo en la decada de los 70's ?
actrices = elenco[(elenco.type == 'actress')&(elenco.year//10 == 197)]
len(actrices)
```

```python
# ¬øCuantos papeles protagonicos existen desde el principio de
# la historia filmatografica hasta 1970 ?

protagonistas = elenco[(elenco.n ==1)&(elenco.year<=1970)]
len(protagonistas)
```
