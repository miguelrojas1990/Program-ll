# Scikit-Learn API

url = [https://arxiv.org/abs/1309.0238](https://arxiv.org/abs/1309.0238)

## Principios generales

* **Consistencia.** Todos los objetos (básicos o compuestos) comparten una interfaz consistente compuesta por un conjunto limitado de métodos. Esta interfaz se documenta de manera consistente para todos los objetos.

* **Inspección.** Los parámetros del constructor y los valores de los parámetros determinados por los algoritmos de aprendizaje se almacenan y se muestran como atributos públicos.

* **No proliferación de clases.** Los algoritmos de aprendizaje son los únicos objetos que se representan mediante clases personalizadas. Los conjuntos de datos se representan como matrices NumPy o matrices dispersas SciPy. Los nombres y valores de los hiperparámetros se representan como cadenas o números de Python estándar siempre que sea posible. Esto mantiene scikitlearn fácil de usar y fácil de combinar con otras bibliotecas.

* **Composición.** Muchas tareas de aprendizaje automático se pueden expresar como secuencias o combinaciones de transformaciones de datos. Algunos algoritmos de aprendizaje también se ven naturalmente como metaalgoritmos parametrizados con otros algoritmos. Siempre que sea viable, se implementan y componen tales algoritmos a partir de bloques de construcción existentes.

* **Valores predeterminados sensatos.** Cada vez que una operación requiere un parámetro definido por el usuario, la biblioteca define un valor predeterminado apropiado. El valor predeterminado debe hacer que la operación se realice de una manera sensata (proporcionando una solución base para la tarea a mano).


# Basic Steps of Using Scikit-Learn API

1. Choose a class of model
2. Choose model hyperparameters
3. Arrage data into features matrix and target array
4. Fit model to data
5. Apply trained model to new data



# Supervised Learning: Simple Linear Regression

```python
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(100)
```

```python
y = 3 * x + np.random.randn(100)
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(x, y);
```

## Step 1. Choose a class of model

```python
from sklearn.linear_model import LinearRegression
```

## Step 2. Choose model hyperparameters

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

## Step 3. Arrage data into features matrix and target array


```python
X = x.reshape(-1, 1)
X.shape
```

## Step 4. Fit model to data


```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

## Step 5. Apply trained model to new data

```python
x_fit = np.linspace(-1, 11)
```

```python
X_fit = x_fit.reshape(-1,1)
```

```python
y_fit = model.predict(X_fit)
```

## Visualise

```python
plt.figure(figsize = (10, 8))
plt.scatter(x, y)
plt.plot(x_fit, y_fit);
```
